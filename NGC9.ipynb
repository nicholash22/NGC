```python
%matplotlib inline
import random
```

```python
#buat list 2 element (head and tail)
head_tail = [0,0]
```

```python
#loop through 50000 trials
trials = 50000
trial = 0
while trial < trials:
    trial = trial + 1
    #get a random 0 or 1
    toss = random.randint(0,1)
    #increment the list element corresponding to the toss result
    head_tail[toss] = head_tail[toss] + 1
    
print (head_tail)
```

```python
# show a pie chart of the result
from matplotlib import pyplot as plt
plt.figure(figsize=(5,5))
plt.pie(head_tail, labels=['head', 'tail'])
plt.legend()
plt.show()
```
